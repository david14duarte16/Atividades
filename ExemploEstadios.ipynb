```python
# Estabelecendo um ponto central do mapa (Centro do Brasil).
center_lat = -14.24007
center_lon = -53.18050
```

```python
# todos os dados dos estadios na variável consulta
consulta = list(estadios.find())
```

```python
# Converte em um data frame
df_convertido = pd.DataFrame(consulta)
```

```python
df_convertido.iloc[0]
```

```python
# criar um mapa a partir do ponto central com 800x600
locationmap = folium.Map(location=[ center_lat, center_lon], zoom_star = 2, width=800, height=600)
```

```python
qtde = len(df_convertido)
```

```python
qtde
```

```python
df_convertido = df_convertido.dropna(subset=['longitude', 'latitude'])
for i in range(len(df_convertido)):
    lng = float(df_convertido.iloc[i]['longitude'])
    lat = float(df_convertido.iloc[i]['latitude'])
    description = 'Estádio: ' + str(df_convertido.iloc[i]['nome']) + ' Capacidade: ' + str(df_convertido.iloc[i]['capacidade'])
    folium.Marker([lat, lng],
                  popup=description,
                  icon=folium.Icon(color='red')).add_to(locationmap)
```
